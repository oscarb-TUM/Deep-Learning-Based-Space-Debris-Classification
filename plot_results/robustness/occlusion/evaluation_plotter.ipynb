```python
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Load JSON data
def load_json_data(filepath):
    with open(filepath, 'r') as file:
        return json.load(file)

subsampling_lstm_data = load_json_data('subsampling_LSTM_vanilla_db.json')['data']
sampling_lstm_bi_data = load_json_data('sampling_lstm_bi_raw.json')['data']
sampling_transformer_data = load_json_data('sampling_transformer_full_raw.json')['data']

# Extract intensities and accuracies
def extract_data(data):
    intensities = [item[0] for item in data]
    accuracies = [item[1] for item in data]
    return intensities, accuracies

intensities_lstm, accuracies_lstm = extract_data(subsampling_lstm_data)
intensities_lstm_bi, accuracies_lstm_bi = extract_data(sampling_lstm_bi_data)
intensities_transformer, accuracies_transformer = extract_data(sampling_transformer_data)

# Set the font
plt.rcParams.update({'font.family': 'Palatino', 'font.size': 10})

# Set seaborn color palette
colors = sns.color_palette("tab10")

# Plot
fig, ax = plt.subplots(figsize=(6, 2))
ax.plot(intensities_lstm, accuracies_lstm, label='DB: LSTM', color=colors[0])
ax.plot(intensities_lstm_bi, accuracies_lstm_bi, label='Raw: Bi-LSTM', color=colors[1])
ax.plot(intensities_transformer, accuracies_transformer, label='Raw: FullTRAN', color=colors[2])

# Adding axis labels
ax.set_xlabel('Intensity')
ax.set_ylabel('Accuracy')

# Add legend to the plot
ax.legend(loc='best', fontsize='small', fancybox=True, framealpha=0.5)

plt.tight_layout()

plt.savefig('/Users/oscarb/Downloads/random_dropout_metrics.pdf', format='pdf')
plt.show()
```

```python
# Create a new figure for the legend
fig_legend, ax_legend = plt.subplots(figsize=(2, 2))
# Create empty plots for each legend item
legend_lines = [plt.Line2D([0], [0], color=colors[i], linewidth=3, label=label) for i, label in enumerate(['LSTM Vanilla', 'LSTM BI', 'Transformer'])]

# Create the legend
ax_legend.legend(handles=legend_lines, loc='center')
ax_legend.axis('off')  # Hide the axes

plt.show()
```
