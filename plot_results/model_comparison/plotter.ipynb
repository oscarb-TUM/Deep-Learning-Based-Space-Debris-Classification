```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Palatino']
plt.rcParams['font.size'] = 9 
```

```python

# Model name
model = 'Bi-LSTM'

# Data
epochs = ['10', '30']
accuracy = [94.72, 95.74]

# Set the width of the bars
barWidth = 0.4

# Set position of bar on X axis
r1 = np.arange(len(epochs))

# Make the plot
plt.figure(figsize=(2, 2))
plt.bar(r1, accuracy, color='#9467bd', width=barWidth, edgecolor='grey', label='Accuracy') # Blue color

# Add xticks
plt.xlabel('Epoch', fontweight='bold')
plt.ylabel('Percentage', fontweight='bold')
plt.xticks([r for r in range(len(epochs))], epochs)
# plt.ylim(90, 97.5)  # Zoom in on the crucial range
plt.xlim(-0.5, len(epochs)-0.5)  # Adjust x-axis limits

# Create legend & Show graphic

plt.title(f'{model}: Accuracy')
plt.tight_layout()

plt.savefig("/Users/oscarb/Downloads/bar_LSTM_acc_30.pdf", format='pdf')
plt.show()
```

```python

# Load your data

file_path = '/Users/oscarb/Bachelor_Thesis/DL_Pipeline/repo/synthetic-data/results/plot_data/fullTRAN_loss_50_raw.csv'  # Make sure to replace this with your actual file path
#file_path = '/Users/oscarb/Bachelor_Thesis/DL_Pipeline/repo/synthetic-data/results/plot_data/lstm_loss_epoch=30.csv'  # Make sure to replace this with your actual file path
data = pd.read_csv(file_path)

# Extracting relevant columns for plotting
epochs = data['epoch']

#training_loss = data['LSTM_bidirectional_long_run - loss/train_epoch']
#validation_loss = data['LSTM_bidirectional_long_run - loss/val_epoch']

training_loss = data['FAVORITE_transformer_db_false_full_attention - loss/train_epoch']
validation_loss = data['FAVORITE_transformer_db_false_full_attention - loss/val_epoch']

# Adjusting font and figure size to match LaTeX settings
plt.rcParams['font.family'] = 'serif'  # 'mathpazo' font approximation
plt.rcParams['font.serif'] = ['Palatino']  # You might need to adjust this based on available fonts
plt.rcParams['font.size'] = 9  # Adjust based on your document's font size
figure_width = 2  # Approx a third of the textwidth, in inches
figure_height = 2  # Height in inches, adjust as needed

# Plotting training and validation loss
plt.figure(figsize=(figure_width, figure_height))
plt.plot(epochs, training_loss, color='#0074D9', label='train')  # Orange color
plt.plot(epochs, validation_loss, color='#FF5733', label='val', linestyle='--')  # Blue color

# Labeling the axes and creating a legend
plt.xlabel('Epoch', fontweight='bold')
plt.ylabel('Loss', fontweight='bold')
plt.title('FullTRAN - Raw')
plt.legend()

plt.xticks(range(min(epochs), max(epochs)+1, 10))

# Adjusting plot aesthetics
plt.tight_layout()

# Optionally save the figure to a file
plt.savefig("/Users/oscarb/Downloads/raw_FullTRAN_50_loss_curve.pdf", format='pdf')

plt.show()
```
