```python
import matplotlib.pyplot as plt
import numpy as np

models = ['MLP', 'ResNet', 'LSTM', 'Bi-LSTM', 'FullTRAN', 'SparseTRAN']
raw_scores = np.array([0.6772, 0.8344, 0.7034, 0.9496, 0.9393, 0.8281])
db_scores = np.array([0.803, 0.9476, 0.9496, 0.9472, 0.8984, 0.9329])
differences = db_scores - raw_scores

plt.figure(figsize=(6, 2))
plt.bar(models, differences, color='purple')
plt.ylabel('Difference (dB - Raw)')
plt.title('Difference Between dB and Raw Scores')
plt.tight_layout()
plt.show()
```

```python

# Model names
models = ['MLP', 'ResNet', 'LSTM', 'Bi-LSTM', 'FullTRAN', 'SparseTRAN']

# Data
raw_scores = [0.6772, 0.8344, 0.7034, 0.9496, 0.9393, 0.8281]
db_scores = [0.803, 0.9476, 0.9496, 0.9472, 0.8984, 0.9329]

inverted_scores_raw = [1 - score for score in raw_scores]
inverted_scores_db = [1 - score for score in db_scores]

# Set the width of the bars
barWidth = 0.3

# Set position of bar on X axis
r1 = np.arange(len(models))
r2 = [x + barWidth + 0.1 for x in r1]

# Make the plot
plt.figure(figsize=(6, 2))
plt.bar(r1, inverted_scores_raw, color='#4C72B0', width=barWidth, edgecolor='grey', label='Raw') # Blue color
plt.bar(r2, inverted_scores_db, color='#55A868', width=barWidth, edgecolor='grey', label='dB') # Green color

# Add xticks on the middle of the group bars
#plt.xlabel('Model', fontweight='bold')
# plt.ylabel('Error Percentage', fontweight='bold')
plt.ylabel('Error Percentage')
plt.xticks([r + barWidth/2 + 0.05 for r in range(len(models))], models, rotation=0)

#plt.title("Error Level")
# Set y-axis limits
# plt.ylim(0.25, 1.0)

# Create legend & Show graphic
plt.legend()
plt.tight_layout()

plt.savefig('/Users/oscarb/Downloads/new_correct_all_comp_raw_db.pdf', format='pdf', bbox_inches='tight')
plt.show()
```
