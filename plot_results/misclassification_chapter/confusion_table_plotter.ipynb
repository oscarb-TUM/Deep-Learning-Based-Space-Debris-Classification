```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import rcParams

# Set the font to Palatino, size 9 for the entire plot
rcParams['font.family'] = 'Palatino'
rcParams['font.size'] = 9

# Load the CSV file
df = pd.read_csv('confusion_lstm.csv')  # Update this to the path of your CSV file

# Pivot the DataFrame to create a confusion matrix format
confusion_matrix = df.pivot(index="Actual", columns="Predicted", values="nPredictions").fillna(0)

# Plot the confusion matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix, annot=True, fmt='g', cmap='Blues', cbar=False)

# Title and labels
plt.title('Confusion Matrix', fontsize=12)
plt.xlabel('Predicted Categories', fontsize=10)
plt.ylabel('Actual Categories', fontsize=10)

# Display the plot
plt.tight_layout()
plt.show()
```
